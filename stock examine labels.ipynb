```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ZIONO_weekly_return_volatility.csv")
df.head()
```

```python
df_21 = df[df.Year==2021].copy()
df_21.label.fillna(method="ffill", inplace=True)
df_21.set_index("Week_Number", inplace=True)
df_21.head(10)
```

```python
df_22 = df[df.Year==2022].copy()
df_22.label.fillna(method="ffill", inplace=True)
df_22.set_index("Week_Number", inplace=True)
df_22.head(15)
```

```python
# plt.figure(figsize=(30, 18))
plt.rcParams['figure.figsize'] = (14.0, 8.0)
colors = []
texts = []
for i in range(1, len(df_21.label) + 1):
    #     print(df_21.label[i])
    colors.append('#FF0000' if df_21.label[i] == 'red' else '#00FF00')
    texts.append(f"{i+1}")

df_21.plot.scatter("mean_return", "volatility", s=24, c=colors)
for i in range(len(texts)):
    plt.annotate(texts[i], xy = (df_21.mean_return[i+1], df_21.volatility[i+1]), xytext=(df_21.mean_return[i+1]+0.01, df_21.volatility[i+1]+0.01))
plt.show()
```

```python
colors = []
for i in range(1, len(df_22.label) + 1):
    #     print(df_21.label[i])
    colors.append('#FF0000' if df_22.label[i] == 'red' else '#00FF00')

df_22.plot.scatter("mean_return", "volatility", s=18, c=colors)

for i in range(len(texts)):
    plt.annotate(texts[i], xy = (df_22.mean_return[i+1], df_22.volatility[i+1]), xytext=(df_22.mean_return[i+1]+0.01, df_22.volatility[i+1]+0.01))

plt.show()
```
