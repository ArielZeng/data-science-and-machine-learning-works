```python
#5
# zinc = 11, B6 = 1.3, iron=8, vitamin c = 90
df_in_range = df[df.Category.isin(['BEEF','PORK','CHEESE','OEREALS','LAMB'])]
df_in_range.head()
```

```python
df_in_range = df_in_range[['Data.Vitamins.Vitamin B6','Data.Major Minerals.Zinc','Data.Major Minerals.Iron','Data.Major Minerals.Calcium']]
df_in_range
```

```python
# zinc = 11, B6 = 1.3, iron=8, calcium =50
standard = {'Data.Vitamins.Vitamin B6':0.5,'Data.Major Minerals.Zinc':5,'Data.Major Minerals.Iron':4,'Data.Major Minerals.Calcium':50}
stand = {}
for ele in standard.items():
  stand[ele[0]]=ele[1]*0.25
standard = stand
standard
```

```python
# del df_in_range['label']
def get_label(x):
  is_good = 1
  for ind in x.index:
   if x[ind]< standard[ind]:
     is_good = 0
     break
  return is_good
df_in_range['label'] = df_in_range[df_in_range.columns].apply(lambda x: get_label(x), axis=1)
df_in_range.head()
```

```python
df_train = df_in_range.iloc[:700]
df_test = df_in_range.iloc[701:]
```

```python
#del ['label']
def get_predict(x):
  return 0 if x['Data.Vitamins.Vitamin B6']<0.5 and x['Data.Major Minerals.Zinc']<4.5 else 1
#7
df_test['predict_label'] = df_test[df_test.columns].apply(lambda x: get_predict(x), axis=1)
df_test.head()
```

```python
#8
def accuracy(x):
  return 1 if x['label']== x['predict_label']  else 0
df_test['accuracy'] = df_test[df_test.columns].apply(lambda x: accuracy(x), axis=1)
df_test.head()
```

```python
def TP(x):
  return 1 if x['predict_label']==1 and x['label']==1  else 0
def FP(x):
  return 1 if x['predict_label']==1 and x['label']==0  else 0
def TN(x):
  return 1 if x['predict_label']==0 and x['label']==0  else 0
def FN(x):
  return 1 if x['predict_label']==0 and x['label']==1  else 0
df_test['TP'] = df_test[df_test.columns].apply(lambda x: TP(x), axis=1)
df_test['FP'] = df_test[df_test.columns].apply(lambda x: FP(x), axis=1)
df_test['TN'] = df_test[df_test.columns].apply(lambda x: TN(x), axis=1)
df_test['FN'] = df_test[df_test.columns].apply(lambda x: FN(x), axis=1)
df_test.head()
```

```python
tp_count = df_test.TP.sum()#ç®—tpæ¬¡æ•°
fn_count = df_test.FN.sum()
tn_count = df_test.TN.sum()
fp_count = df_test.FP.sum()
acc_count = df_test.accuracy.sum()#times of guess accurately
total_count = df_test.shape[0]
```

```python
tp_ratio = tp_count/(tp_count+fn_count)
tn_ratio = tn_count/(tn_count + fp_count)
acc_ratio = acc_count/total_count
[tp_count,fn_count,tn_count,fp_count,acc_ratio, tp_ratio, tn_ratio]
```

```python
from sklearn.metrics import roc_auc_score

auc_score = roc_auc_score(df_test.label, df_test.predict_label)
print(auc_score)
```
