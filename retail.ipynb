```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("online_retail.csv")
df.head()
```

```python
def get_digit(x):
    s = f"{x}"
    return int(s[0])

    

df["digit"] = df.UnitPrice.apply(lambda x:get_digit(x))
df.head(30)
```

```python
df_in_range = df[df.digit != 0]
df_in_range.head(30)
```

```python
freq = df_in_range.digit.value_counts()
freq
```

```python
total_freq = sum(freq)
total_freq
```

```python
freq = freq / total_freq
freq
```

```python
uniform_freq = pd.Series([1, 1, 1, 1, 1, 1, 1, 1, 1])
uniform_freq = uniform_freq / 9
uniform_freq
```

```python
import math
bernford_p = pd.Series([1, 2, 3, 4, 5, 6, 7, 8, 9])
bernford_p = bernford_p.map(lambda d:math.log10(1+1/d))
bernford_p
```

```python
df_digit = pd.DataFrame({"real":freq, "uniform":uniform_freq, "bern":bernford_p})
df_digit
```

```python
df_digit.real = df_digit.real.shift(-1)
df_digit = df_digit[:-1]
df_digit
```

```python
plt.bar(df_digit.index+1, df_digit['bern'], width=0.2, color='r', label='bern')     
plt.bar(df_digit.index+1+0.2, df_digit['real'], width=0.2, color ='b', label='real')   
plt.bar(df_digit.index+1+0.4, df_digit['uniform'], width=0.2, color ='g', label='uniform')   
plt.xticks(df_digit.index+1+0.2, df_digit.index+1)  
plt.title("freq comparison")
plt.legend()
plt.show()
```

```python
df_relative_err = df_digit.copy()
df_relative_err["bern_err"] = df_relative_err.bern - df_relative_err.real
df_relative_err["uniform_err"] = df_relative_err.uniform - df_relative_err.real
df_relative_err
```

```python
plt.bar(df_relative_err.index+1, df_relative_err['bern_err'], width=0.2, color='r', label='bern_err')   
plt.bar(df_relative_err.index+1+0.2, df_relative_err['uniform_err'], width=0.2, color ='g', label='uniform_err')   
plt.xticks(df_relative_err.index+1+0.2, df_relative_err.index+1)  
plt.title("relative error")
plt.legend(loc = 4)
plt.show()
```

```python
df_relative_err.bern_err.std()
```

```python
df_relative_err.uniform_err.std() # we can see bernford has smaller RMSE
```

```python
df_jp = df_in_range[df_in_range.Country == "Japan"]
freq_jp = df_jp.digit.value_counts()
total_jp = sum(freq_jp)
freq_jp = freq_jp / total_jp
freq_jp
```

```python
df_Isreal = df_in_range[df_in_range.Country == "Israel"]
freq_is = df_Isreal.digit.value_counts()
total_is = sum(freq_is)
freq_is = freq_is / total_is
freq_is
```

```python
df_countries = pd.DataFrame({"fr_f": freq_fr, "jp_f":freq_jp, "Is_f":freq_is, "P_uniform":uniform_freq, "PI_bern":bernford_p})
df_countries.P_uniform = df_countries.P_uniform.shift()
df_countries.PI_bern = df_countries.PI_bern.shift()
df_countries = df_countries[1:]
df_countries
```

```python
df_countries['Is_f_err'] = df_countries.Is_f - df_countries.P_uniform
df_countries['jp_f_err'] = df_countries.jp_f - df_countries.P_uniform
df_countries['fr_f_err'] = df_countries.fr_f - df_countries.P_uniform
df_countries
```
