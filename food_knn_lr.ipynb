```python
#5
# zinc = 11, B6 = 1.3, iron=8, vitamin c = 90
df_in_range = df[df.Category.isin(['BEEF','PORK','CHEESE','OEREALS','LAMB'])]
df_in_range.head()
```

```python
df_in_range = df_in_range[['Data.Vitamins.Vitamin B6','Data.Major Minerals.Zinc','Data.Major Minerals.Iron','Data.Major Minerals.Calcium']]
df_in_range
```

```python
# zinc = 11, B6 = 1.3, iron=8, calcium =50
standard = {'Data.Vitamins.Vitamin B6':0.5,'Data.Major Minerals.Zinc':5,'Data.Major Minerals.Iron':4,'Data.Major Minerals.Calcium':50}
stand = {}
for ele in standard.items():
  stand[ele[0]]=ele[1]*0.25
standard = stand
standard
```

```python
# del df_in_range['label']
def get_label(x):
  is_good = 1
  for ind in x.index:
   if x[ind]< standard[ind]:
     is_good = 0
     break
  return is_good
df_in_range['label'] = df_in_range[df_in_range.columns].apply(lambda x: get_label(x), axis=1)
df_in_range.head()
```

```python
df_in_range[df_in_range.label==1]
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import pandas as pd



# # Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    df_in_range.iloc[:, 0:4], df_in_range.iloc[:, 4], test_size=0.3, random_state=434343)


clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred)
acc
```
