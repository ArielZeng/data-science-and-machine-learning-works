```python
# 7 
df.shape

day_num = 0
prev_day = None
inday_index = 1
def index_day(new_day):
    global day_num, prev_day, inday_index
#     print(new_day)
    if prev_day is None:
        prev_day = new_day
        day_num += 1
        inday_index = 1
    elif prev_day == new_day:
        inday_index += 1
        pass
    else:
        prev_day = new_day
        day_num += 1
        inday_index = 1
        
    return day_num* 1000 + inday_index
    


df["day_tag"] = df.day.apply(index_day)
df["day_NO"] = df.day_tag // 1000
df["inday_index"] = df.day_tag % 1000
min_dayNo = df.day_NO.min()
max_dayNo = df.day_NO.max()
min_dayNo, max_dayNo
for cur_day in range (min_dayNo, max_dayNo + 1):
    df_day = df[df.day_NO == cur_day]
    print(f"day = {cur_day} corr between index and tip", df_day[["inday_index", "tip"]].corr().iloc[0, 1])
    
# from the table below, we can see some times the corr is positive, which 
# means tips are increasing with time in that day. While sometimes, the corr is negtive, which
# means tips are decreasing with time in this day.
```
